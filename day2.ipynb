```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class linkedlist:
    def __init__(self):
        self.head = None

    def insertionatend(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def merginglinkedlists(self, head1, head2):
        temp = None
        if head1 is None:
            return head2
        elif head2 is None:
            return head1
        elif head1.data <= head2.data:
            temp = head1
            temp.next = self.merginglinkedlists(head1.next, head2)
        else:
            temp = head2
            temp.next = self.merginglinkedlists(head1, head2.next)
        return temp

    def printlist(self):
        temp = self.head
        while temp:
            print(str(temp.data), end=" ")
            temp = temp.next

llist1 = linkedlist()
llist1.insertionatend(1)
llist1.insertionatend(3)
llist1.insertionatend(5)
llist1.printlist()
print()
llist2 = linkedlist()
llist2.insertionatend(2)
llist2.insertionatend(4)
llist2.insertionatend(6)
llist2.printlist()
print()
llist3 = linkedlist()
llist3.head = llist3.merginglinkedlists(llist1.head, llist2.head)
llist3.printlist()
```

```python
class Stack:
    def __init__(self):
        self.items = []
    def push(self, item):
        self.items.append(item)
        print(item, "pushed into stack")
    def pop(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.items.pop()
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
            return None
        return self.items[-1]
    def is_empty(self):
        return len(self.items) == 0

    # Display stack
    def display(self):
        print("Stack:", self.items)
s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.display()

print("Top element:", s.peek())
print("Popped:", s.pop())
s.display()
```

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
#stack class using linked list
class Stack:
  def __init__(self):
    self.top=None
  #push operation
  def push(self,data):
    new_node=Node(data)
    new_node.next=self.top
    self.top=new_node
    print(data,"pushed into stack")
  #pop method
  def pop(self):
    if self.is_empty():
      print("Stack is empty")
      return None
    popped_value=self.top.data
    self.top=self.top.next
    return popped_value

  def peek(self):
    if self.is_empty():
      print("stack is empty")
      return None
    return self.top.data
  def is_empty(self):
    return self.top is None
  def display(self):
    if self.is_empty():
      print("stack is empty")
      return
    temp=self.top
    print("stack elements(top to botom):")
    while temp:
      print(temp.data)
      temp=temp.next
s=Stack() # Changed from s=stack() to s=Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.push(50)
print()
s.display()
print("top element:",s.peek())
print("popped:",s.pop())
s.display()
print()
s.pop()
s.pop()
s.pop()
s.pop()
s.pop()
s.is_empty()
```

```python
def sum_even(n):
    if n == 2:
        return 2
    if n % 2 == 0:
        return n + sum_even(n - 1)
    else:
        return sum_even(n - 1)

num = int(input("Enter a number: "))
print("Sum of even numbers:", sum_even(num))
```
